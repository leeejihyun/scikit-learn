# 6장. 결정 트리

## 6.1 결정 트리 학습과 시각화

#### 결정 트리 학습

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris() # 데이터 로드
X = iris.data[:, 2:] # 꽃잎의 길이와 너비
y = iris.target # 꽃 종류를 정답값 y로 설정

tree_clf = DecisionTreeClassifier(max_depth=2) # 깊이 2인 결정 트리 분류기 생성
tree_clf.fit(X, y) # 데이터 X와 정답값 y를 사용하여 분류기 학습
```

```python
from sklearn.tree import export_graphviz

export_graphviz(
tree_clf,
out_file="iris_tree.dot",
feature_names=iris.feature_names[2:],
class_names=iris.target_names,
rounded=True, # rounded=True: draw node boxes with rounded corners and use Helvetica fonts instead of Times-Roman
filled=True # filled=True: paint nodes to indicate majority class for classification, extremity of values for regression, or purity of node for multi-output
)
```
